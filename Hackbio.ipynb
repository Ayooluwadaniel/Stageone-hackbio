```python
def analyze_dna(dna_sequences):

    gc_contents = []
    for dna in dna_sequences:
        g_content = dna.count('G')
        c_content = dna.count('C')
        gc_content_percentage = ((g_content + c_content) / len(dna)) * 100
        gc_contents.append(gc_content_percentage)

        print(f"Sequence: {dna} | GC Content: {gc_content_percentage:.2f}%")

    return gc_contents

dna_sequences = ["ATGCGTAC", "GCTTAGCT", "CGTACGTA"]
gc_values = analyze_dna(dna_sequences)
print(gc_values)
```

```python
from scipy import stats

# Sample data: OD600 readings of bacterial growth
control = [0.9, 1.1, 1.2, 1.0, 0.95]
treated = [1.3, 1.5, 1.45, 1.4, 1.35]

# Perform an independent t-test
t_stat, p_value = stats.ttest_ind(control, treated)

print(f"T-statistic: {t_stat}")
print(f"P-value: {p_value}")

# Interpretation
if p_value < 0.05:
    print("The difference in bacterial growth is statistically significant.")
else:
    print("The difference in bacterial growth is not statistically significant.")
```

### DNA to Protein Translation Function

This function will take a DNA sequence and translate it into an amino acid sequence (protein) based on the standard genetic code. It will read the DNA in codons (groups of three bases) and map each codon to its corresponding amino acid.

```python
def dna_to_protein(dna_sequence):
    """
    Translates a DNA sequence into a protein sequence.
    """
    genetic_code = {
        'ATA': 'I', 'ATC': 'I', 'ATT': 'I', 'ATG': 'M',
        'ACA': 'T', 'ACC': 'T', 'ACG': 'T', 'ACT': 'T',
        'AAC': 'N', 'AAT': 'N', 'AAA': 'K', 'AAG': 'K',
        'AGC': 'S', 'AGT': 'S', 'AGA': 'R', 'AGG': 'R',
        'CTA': 'L', 'CTC': 'L', 'CTG': 'L', 'CTT': 'L',
        'CCA': 'P', 'CCC': 'P', 'CCG': 'P', 'CCT': 'P',
        'CAC': 'H', 'CAT': 'H', 'CAA': 'Q', 'CAG': 'Q',
        'CGA': 'R', 'CGC': 'R', 'CGG': 'R', 'CGT': 'R',
        'GTA': 'V', 'GTC': 'V', 'GTG': 'V', 'GTT': 'V',
        'GCA': 'A', 'GCC': 'A', 'GCG': 'A', 'GCT': 'A',
        'GAC': 'D', 'GAT': 'D', 'GAA': 'E', 'GAG': 'E',
        'GGA': 'G', 'GGC': 'G', 'GGG': 'G', 'GGT': 'G',
        'TCA': 'S', 'TCC': 'S', 'TCG': 'S', 'TCT': 'S',
        'TTC': 'F', 'TTT': 'F', 'TTA': 'L', 'TTG': 'L',
        'TAC': 'Y', 'TAT': 'Y', 'TAA': '_', 'TAG': '_',
        'TGC': 'C', 'TGT': 'C', 'TGA': '_', 'TGG': 'W',
    }

    protein_sequence = []
    # Ensure sequence length is a multiple of 3
    if len(dna_sequence) % 3 != 0:
        print("Warning: DNA sequence length is not a multiple of 3. Truncating.")
        dna_sequence = dna_sequence[:-(len(dna_sequence) % 3)]

    for i in range(0, len(dna_sequence), 3):
        codon = dna_sequence[i:i+3]
        amino_acid = genetic_code.get(codon.upper(), 'X') # 'X' for unknown codon
        protein_sequence.append(amino_acid)

        # Stop translation if a stop codon is encountered
        if amino_acid == '_':
            break

    return "".join(protein_sequence)

# Example usage:
dna_seq = "ATGGCCATTGTAAATAA"
protein_seq = dna_to_protein(dna_seq)
print(f"DNA Sequence: {dna_seq}")
print(f"Protein Sequence: {protein_seq}")

dna_seq_long = "ATGCTATTGATCAAGTAG"
protein_seq_long = dna_to_protein(dna_seq_long)
print(f"\nDNA Sequence: {dna_seq_long}")
print(f"Protein Sequence: {protein_seq_long}")
```

### Hamming Distance Calculation Function

This function calculates the Hamming distance between two strings. The Hamming distance is the number of positions at which the corresponding symbols are different. It is typically used for strings of equal length.

```python
def hamming_distance(s1, s2):
    """
    Calculates the Hamming distance between two strings.
    Returns -1 if strings are of different lengths.
    """
    if len(s1) != len(s2):
        print("Error: Strings must be of the same length to calculate Hamming distance.")
        return -1

    distance = 0
    for i in range(len(s1)):
        if s1[i] != s2[i]:
            distance += 1
    return distance

# Your Slack username and Twitter/X handle
slack_username = "AyooluwaDaniel"
twitter_handle = "Theayooodaniel"

dist = hamming_distance(slack_username, twitter_handle)

if dist != -1:
    print(f"String 1: {slack_username}")
    print(f"String 2: {twitter_handle}")
    print(f"Hamming Distance: {dist}")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.feature_selection import RFE
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error
import pandas as pd

# Define a range of features to select
# The original dataset has 200 features, so we'll test from 10 to 200 with a step of 20
n_features_range = range(10, X_train.shape[1] + 1, 20)

# Lists to store MSE values for different models
mse_linear_list = []
mse_rf_list = []

print("Calculating MSE for different numbers of features...")

for n_features in n_features_range:
    # Perform RFE with the current number of features
    # Using a Linear Regression estimator for RFE as in previous steps
    rfe = RFE(estimator=LinearRegression(), n_features_to_select=n_features, step=10)
    rfe.fit(X_train, y_train)
    selected_features = X_train.columns[rfe.support_]

    # Prepare datasets with selected features
    X_train_selected = X_train[selected_features]
    X_test_selected = X_test[selected_features]

    # --- Linear Regression Model ---
    linear_model = LinearRegression()
    linear_model.fit(X_train_selected, y_train)
    y_pred_linear = linear_model.predict(X_test_selected)
    mse_linear = mean_squared_error(y_test, y_pred_linear)
    mse_linear_list.append(mse_linear)

    # --- Random Forest Regressor Model ---
    # Using n_estimators=90 and random_state=42 as in previous steps
    rf_model = RandomForestRegressor(n_estimators=90, random_state=42)
    rf_model.fit(X_train_selected, y_train)
    y_pred_rf = rf_model.predict(X_test_selected)
    mse_rf = mean_squared_error(y_test, y_pred_rf)
    mse_rf_list.append(mse_rf)
    print(f"  Features: {n_features}, Linear MSE: {mse_linear:.2f}, RF MSE: {mse_rf:.2f}")

# Plotting the results
plt.figure(figsize=(12, 7))
plt.plot(list(n_features_range), mse_linear_list, marker='o', linestyle='-', color='blue', label='Linear Regression MSE')
plt.plot(list(n_features_range), mse_rf_list, marker='x', linestyle='--', color='red', label='Random Forest MSE')

plt.title('Mean Squared Error vs. Number of Selected Features', fontsize=16)
plt.xlabel('Number of Features Selected', fontsize=12)
plt.ylabel('Mean Squared Error (MSE)', fontsize=12)
plt.legend(fontsize=10)
plt.grid(True, linestyle='--', alpha=0.7)
plt.xticks(list(n_features_range))
plt.tight_layout()
plt.show()

print("Plot generated successfully.")
```

```python
import pandas as pd

# Load the dataset
df = pd.read_csv('https://raw.githubusercontent.com/HackBio-Internship/2025_project_collection/refs/heads/main/Python/Dataset/plant_microbe_interaction_data_v2.csv')

# Display the first few rows
print(df.head())
```

```python
from sklearn.preprocessing import StandardScaler

# Initialize the scaler
scaler = StandardScaler()

# Fit and transform the data
scaled_data = scaler.fit_transform(df)
print(scaled_data)
```

```python
from sklearn.cluster import KMeans

# Initialize KMeans with 3 clusters
kmeans = KMeans(n_clusters=3, random_state=42)

# Fit the model
kmeans.fit(scaled_data)

# Add the cluster labels to the original dataframe
df['Cluster'] = kmeans.labels_
```

```python
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA

# Reduce dimensions to 2D for visualization
pca = PCA(n_components=2)
principal_components = pca.fit_transform(scaled_data)

# Create a DataFrame with the principal components
pc_df = pd.DataFrame(data=principal_components, columns=['PC1', 'PC2'])
pc_df['Cluster'] = df['Cluster']

# Plot the clusters
plt.figure(figsize=(10, 6))
for cluster in pc_df['Cluster'].unique():
    cluster_data = pc_df[pc_df['Cluster'] == cluster]
    plt.scatter(cluster_data['PC1'], cluster_data['PC2'], label=f'Cluster {cluster}')
plt.title('K-Means Clustering of Plant-Microbe Interaction Data')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.legend()
plt.show()
```

```python
# Calculate the mean values of each feature per cluster
cluster_means = df.groupby('Cluster').mean()
print(cluster_means)
```
